### Part III: Functions

```python
# 1. Functions to convert between kelvin and celsius

def kelvin_to_celsius(k):
    return k - 273.15

def celsius_to_kelvin(c):
    return c + 273.15
```

```python
# 2. Function to convert to Fahrenheit with a keyword argument

def to_fahrenheit(temp, unit='celsius'):
    if unit == 'kelvin':
        temp = kelvin_to_celsius(temp)
    return (temp * 9/5) + 32
```

```python
# 3. Check that the outputs are sensible

print("0°C to F (Should be 32.0):", to_fahrenheit(0)) 
print("273.15K to F (Should be 32.0):", to_fahrenheit(273.15, unit='kelvin'))
print("100°C to F (Should be 212.0):", to_fahrenheit(100, unit='celsius'))
```

```python
# 4. Convert from Fahrenheit with a keyword argument

def from_fahrenheit(f, unit='celsius'):
    celsius = (f - 32) * 5/9
    if unit == 'kelvin':
        return celsius_to_kelvin(celsius)
    return celsius
```

```python
# 5. Convert feet and inches to meters

def height_to_meters(feet, inches):
    # 1 inch = 0.0254 meters, 1 foot = 12 inches
    total_inches = (feet * 12) + inches
    return total_inches * 0.0254

print("5ft 9in in meters (Should be ~1.75):", round(height_to_meters(5, 9), 2))
```

```python
# 6. Convert meters to feet and inches (returning two arguments)

def meters_to_height(meters):
    total_inches = meters / 0.0254
    feet = int(total_inches // 12)          
    inches = round(total_inches % 12, 2)    
    return feet, inches
```

```python
# 7. Verify the "round trip" conversion

check_heights = [1.50, 1.80, 2.00]

for original_meters in check_heights:
    f, i = meters_to_height(original_meters)
    round_trip_meters = height_to_meters(f, i)
    print(f"Original: {original_meters}m -> {f}ft {i}in -> Round trip: {round_trip_meters:.2f}m")
```
